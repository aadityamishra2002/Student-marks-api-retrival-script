```python
import requests
import pandas as pd
from numpy import array
# Replace with your actual API URL (including any API key if needed)
api_url = "https://script.google.com/macros/s/AKfycbyGNo_x44oEm3wNJ2c8QBKjN5yXpVTCjdBYaTwPdOWZ09Q_toDVhLCXwMwplgLniu6z/exec"

# Optional: Add headers for authentication (e.g., API key)
headers = {}  # Replace with your headers dictionary if applicable

# Send a GET request to the API endpoint
response = requests.get(api_url, headers=headers)

# Check for successful response
if response.status_code == 200:
  # Get the JSON data
  data = response.json()
  print("API response (JSON):")
  print(data)  # This will print the formatted JSON data
else:
  print(f"Error accessing API: {response.status_code}")
```

```python
tests_score= data[0][1:]
arr= array(data[1:])
avg_score= arr[:, 1:].astype(int).sum(axis= 0)/1000
print(avg_score)
df= pd.DataFrame({'Test Names': tests_score, 'Average': avg_score});
df.plot.barh('Test Names', 'Average');
```
